```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import japanize_matplotlib
```

```python
with open("../tmp/max_result.pkl", "rb") as pkl:
    max_result = pickle.load(pkl)
with open("../tmp/prop_result.pkl", "rb") as pkl:
    prop_result = pickle.load(pkl)
with open("../tmp/fix_result.pkl", "rb") as pkl:
    fix_result = pickle.load(pkl)
with open("../tmp/prop_raw_result.pkl", "rb") as pkl:
    prop_raw_result = pickle.load(pkl)
```

```python
fix_result
```

```python
print("max:  ", max_result["h0_profit"].sum())
print("fix:  ", fix_result["h0_profit"].sum())
print("prop1:", prop_result["h0_profit"].sum())
print("prop2:", prop_raw_result["h0_profit"].sum())
```

```python
print("max:  ", max_result["h1_profit"].sum())
print("fix:  ", fix_result["h1_profit"].sum())
print("prop1:", prop_result["h1_profit"].sum())
print("prop2:", prop_raw_result["h1_profit"].sum())
```

```python
# 
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(max_result["date"], max_result["h0_profit"], "b")
# ax.plot(fix_result["date"], fix_result["h0_profit"], "r")
ax.plot(prop_result["date"], prop_result["h0_profit"], "g")

fig.show()
```

```python
# 
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(max_result["date"], max_result["h1_profit"], "b")
ax.plot(fix_result["date"], fix_result["h1_profit"], "r")
ax.plot(prop_result["date"], prop_result["h1_profit"], "g")

fig.show()
```

```python
# 
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(prop_result["date"], prop_result["h0_profit"] - prop_raw_result["h0_profit"], "g")

fig.show()
```

```python
# 
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(prop_result["date"], max_result["h0_profit"] - prop_result["h0_profit"], "b", marker=".", label="")
ax.plot(prop_raw_result["date"], max_result["h0_profit"] - prop_raw_result["h0_profit"], "g", marker=".", label="")

ax.set_xlabel("", fontsize=16)
ax.set_ylabel(" []", fontsize=16)

ax.legend(fontsize=16)

ax.set_title(" 1 ", fontsize=16)

xfmt = mdates.DateFormatter("%y/%m/%d")
xloc = mdates.MonthLocator()

ax.xaxis.set_major_locator(xloc)
ax.xaxis.set_major_formatter(xfmt)

fig.savefig("../images/output4.png")

fig.show()
```

```python
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(prop_result["date"], max_result["h1_profit"] - prop_result["h1_profit"], "b", marker=".", label="")
ax.plot(prop_result["date"], max_result["h1_profit"] - prop_raw_result["h1_profit"], "g", marker=".", label="")

ax.set_xlabel("", fontsize=16)
ax.set_ylabel(" []", fontsize=16)

ax.legend(fontsize=16)

ax.set_title(" 2 ", fontsize=16)


xfmt = mdates.DateFormatter("%y/%m/%d")
xloc = mdates.MonthLocator()

ax.xaxis.set_major_locator(xloc)
ax.xaxis.set_major_formatter(xfmt)

fig.savefig("../images/output5.png")

fig.show()
```

```python
# 
fig, ax = plt.subplots()
ax.grid()
fig.set_figwidth(8)
fig.set_figheight(6)

ax.plot(prop_result["date"], max_result["h0"], "b", marker=".", label=" 1")
ax.plot(prop_result["date"], max_result["h1"], "g", marker=".", label=" 2")

xfmt = mdates.DateFormatter("%m/01")
xloc = mdates.MonthLocator()

ax.xaxis.set_major_locator(xloc)
ax.xaxis.set_major_formatter(xfmt)

ax.legend(fontsize=16)

ax.set_xlabel("", fontsize=16)
ax.set_ylabel(" [kg]", fontsize=16)

fig.show()
```
